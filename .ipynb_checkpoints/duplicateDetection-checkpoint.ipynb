```python
# Search for sub-category tops and tunics
df2 = df1.loc[df1['categories'] == "Apparels>Women>Western Wear>Shirts, Tops & Tunics>Tunics" ]

# Changes Indexes as we have subsetted the original dataframe
df2.index = range(len(df2))

# Display the dataframe
df2
```

```python
cols = ['specificationList', 'offers', 'sizeUnit', 'storage', 'displaySize', 'idealFor']

# Dropping the above columns as they are redundant
df2.drop(cols, inplace=True, axis=1)

# Display updated dataframe
df2
```

```python
# Checking the information for our updated Dataframe
df2.info()
```

```python
# There are quite a few NaN's in columns like Description & Neck columns
# We'll replace those with '', so that cosine similarity is computed smoothly
df2['description'] = df2['description'].fillna('')
df2['deliveryTime'] = df2['deliveryTime'].fillna('')
df2['keySpecsStr'] = df2['keySpecsStr'].fillna('')
df2['detailedSpecsStr'] = df2['detailedSpecsStr'].fillna('')
df2['sleeve'] = df2['sleeve'].fillna('')
df2['neck'] = df2['neck'].fillna('')
```

```python
# Checking the dataframe
df2.info()
```

```python
# Checking the first 5 rows of the dataframe
df2.head()
```

```python
# Displaying all the columns of our Dataframe
df2.columns
```

```python
# Indexes corresponding to columns that are required for computing Cosine Similarity
idx = [1,2,3,9,10,16,17,18,19,20,24,25]

# List to store the score for cosine similarity
score = []

# List to store the normalised score(score lies between 0 to 1)
score1 = []

# Lists to create our duplicate products dataframe
l1 = []
l2 = []

# Compute the Cosine Similarity Scores for all the above mentioned columns 
for i in range(0, 50):
    for k in range(1, 50):
        for j in idx:
            if (i != k) and (i<k):
                p1 = nlp(df2.iloc[i, j])
                p2 = nlp(df2.iloc[k, j])
                s = p1.similarity(p2)
                
                score.append(s)
        
for i in range(0, len(score), 12):
    score1.append(sum(score[i:i+12])/12)

for i in range(0, 50):
    for k in range(1, 50):
        if (i != k) and (i<k):
            l1.append(df2.iloc[i,0])
            l2.append(df2.iloc[k,0])

# Columns for new Dataframe
newCols = ['Item-1', 'Item-2', 'Similarity Score']

# New Dataframe that contains the Similiarty scores of the Products
productsDf = pd.DataFrame(columns=newCols)

# Updating our columns with the required values
productsDf['Item-1'] = l1
productsDf['Item-2'] = l2
productsDf['Similarity Score'] = score1

# We set our threshold at 0.80 i.e. Product pairs having a similarity score > 0.80
# are duplicates
duplicateProductList = productsDf.loc[productsDf['Similarity Score'] > 0.80]
duplicateProductList.index = range(len(duplicateProductList))

# Dataframe that displays the list of duplicate products
duplicateProductList
```

```python
# Dictionary that contains list of all duplicate products in required format
new_dict = defaultdict(list)

# my_list = [(i,[a,b]) for i, a,b in zip(duplicateProductList['Item-1'], duplicateProductList['Item-2'], duplicateProductList['Similarity Score'])]
for row in duplicateProductList.itertuples(index=False):
    a, b, c = row
    new_dict[a].append((b,c))
    new_dict[b].append((a,c))
        
# Displaying our Duplicate Products Dictionary        
print(new_dict)
```

```python
# Creating json that contains our Dictionary of duplicate items
with open('duplicateProducts.json', 'w') as fp:
    json.dump(new_dict, fp)
```
